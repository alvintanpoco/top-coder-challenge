```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load and Flatten Data ---
# This part is technically from Hour 1, but necessary for the analysis.
with open("public_cases.json") as f:
    raw_data = json.load(f)

inputs_df = pd.json_normalize([item["input"] for item in raw_data])
expected_df = pd.Series([item["expected_output"] for item in raw_data], name="expected_output")
df = pd.concat([inputs_df, expected_df], axis=1)

# --- Hour 2: Visual Analysis ---

# Set up the plot style for better readability
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
fig.suptitle('Hour 2: Pattern Recognition & Breakpoint Discovery', fontsize=20, weight='bold')

# 1. Scatter Plots to show raw relationships
sns.scatterplot(ax=axes[0, 0], data=df, x="miles_traveled", y="expected_output", alpha=0.5)
axes[0, 0].set_title('Output vs. Miles Traveled', weight='bold')

sns.scatterplot(ax=axes[0, 1], data=df, x="trip_duration_days", y="expected_output", alpha=0.5)
axes[0, 1].set_title('Output vs. Trip Duration', weight='bold')

sns.scatterplot(ax=axes[0, 2], data=df, x="total_receipts_amount", y="expected_output", alpha=0.5)
axes[0, 2].set_title('Output vs. Receipts Amount', weight='bold')


# 2. Bucketed Bar Charts to identify thresholds and non-linearities

# --- Mileage Breakpoints ---
mileage_bins = [0, 100, 200, 300, 400, 500, 700, 1000, 1400]
df['mileage_bin'] = pd.cut(df['miles_traveled'], bins=mileage_bins)
mileage_grouped = df.groupby('mileage_bin', observed=True)['expected_output'].mean()
mileage_grouped.plot(kind='bar', ax=axes[1, 0], color='skyblue')
axes[1, 0].set_title('Mean Output by Mileage Buckets', weight='bold')
axes[1, 0].set_xlabel('Miles Traveled Bins')
axes[1, 0].set_ylabel('Mean Expected Output')
axes[1, 0].tick_params(axis='x', rotation=45)

# --- Duration Breakpoints ---
duration_bins = [0, 1, 3, 5, 7, 10, 14]
df['duration_bin'] = pd.cut(df['trip_duration_days'], bins=duration_bins)
duration_grouped = df.groupby('duration_bin', observed=True)['expected_output'].mean()
duration_grouped.plot(kind='bar', ax=axes[1, 1], color='salmon')
axes[1, 1].set_title('Mean Output by Duration Buckets', weight='bold')
axes[1, 1].set_xlabel('Trip Duration (Days) Bins')
axes[1, 1].set_ylabel('Mean Expected Output')
axes[1, 1].tick_params(axis='x', rotation=45)

# --- Receipts Breakpoints ---
receipts_bins = [0, 500, 1000, 1500, 1800, 2500]
df['receipts_bin'] = pd.cut(df['total_receipts_amount'], bins=receipts_bins)
receipts_grouped = df.groupby('receipts_bin', observed=True)['expected_output'].mean()
receipts_grouped.plot(kind='bar', ax=axes[1, 2], color='lightgreen')
axes[1, 2].set_title('Mean Output by Receipts Buckets', weight='bold')
axes[1, 2].set_xlabel('Total Receipts Amount ($) Bins')
axes[1, 2].set_ylabel('Mean Expected Output')
axes[1, 2].tick_params(axis='x', rotation=45)

# --- Final Touches ---
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# --- Print Final Report based on visuals ---
print("""
=================================================
Hour 2 Analysis Report: Nonlinear Shifts Detected
=================================================

Based on the visual analysis of the bucketed means:

1. Mileage Traveled:
   - The relationship is nonlinear.
   - Key breakpoints appear to exist around 100, 500, and 1000 miles, where the rate of reimbursement changes significantly.

2. Trip Duration:
   - The per-diem reimbursement is clearly tiered, not a flat daily rate.
   - Significant changes in reimbursement rate occur around 3, 7, and 10 days, suggesting a complex policy with plateaus or bonuses.

3. Receipts Amount:
   - The reimbursement appears mostly linear but shows signs of capping out.
   - The growth in reimbursement slows and flattens around the $1,800 mark, indicating a likely cap.

4. The mileage pattern suggests different rate tiers (possibly short trips, medium trips, long trips, and extended travel categories)
   - The duration buckets align with typical corporate policies (weekend trips, week-long, extended assignments)
   The receipts capping appears quite hard - there's very little variation above the $1,800 threshold

Conclusion: A simple linear model will be insufficient. The reimbursement function is definitively a piecewise function with multiple tiers and caps.
""")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load and Flatten ---
df_raw = pd.read_json("public_cases.json")
df_input = pd.json_normalize(df_raw["input"])
df = pd.concat([df_input, df_raw["expected_output"]], axis=1)
df.columns = ['trip_duration_days', 'miles_traveled', 'total_receipts_amount', 'expected_output']

# --- STEP 1: Define f_r(receipts) as min(receipts, 1800) ---
df["f_r"] = df["total_receipts_amount"].clip(upper=1800)

# --- STEP 2: First-order residual (output' = output - f_r) ---
df["output_prime"] = df["expected_output"] - df["f_r"]

# --- STEP 3a: Mileage buckets & mean output' ---
miles_bins = [0, 100, 200, 300, 400, 500, 700, 1000, 1400]
df["miles_bin"] = pd.cut(df["miles_traveled"], bins=miles_bins)
miles_grouped = df.groupby("miles_bin", observed=True)["output_prime"].mean().reset_index()

# --- Compute bin widths from intervals ---
miles_grouped["Œî Miles"] = miles_grouped["miles_bin"].apply(lambda x: x.right - x.left)
miles_grouped["Mean output_prime"] = miles_grouped["output_prime"]
miles_grouped["Œî Output"] = miles_grouped["Mean output_prime"].diff()
miles_grouped["$/mile"] = (miles_grouped["Œî Output"] / miles_grouped["Œî Miles"]).round(4)

# --- Rename for clarity ---
miles_df = miles_grouped[["miles_bin", "Mean output_prime", "Œî Output", "Œî Miles", "$/mile"]]
miles_df.columns = ["Bucket", "Mean output_prime", "Œî Output", "Œî Miles", "$/mile"]

print("üìä Per-mile reimbursement estimate per bin:")
print(miles_df.round(4))

# --- Plot ---
miles_grouped.set_index("miles_bin")["Mean output_prime"].plot(
    kind="bar", figsize=(10, 4), title="output_prime by miles_traveled buckets"
)
plt.ylabel("Mean output_prime")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- STEP 4: Define f_m from observation (placeholder) ---
def f_m(m):
    if m <= 500:
        return m * 0.25
    elif m <= 1000:
        return 500 * 0.25 + (m - 500) * 0.50
    else:
        return 500 * 0.25 + 500 * 0.50 + (m - 1000) * 0.70

df["f_m"] = df["miles_traveled"].apply(f_m)

# --- STEP 5: Second-order residual ---
df["output_double_prime"] = df["output_prime"] - df["f_m"]

# --- STEP 6a: Days buckets & mean output'' ---
day_bins = [0, 1, 3, 5, 7, 10, 14]
df["days_bin"] = pd.cut(df["trip_duration_days"], bins=day_bins)
days_grouped = df.groupby("days_bin", observed=True)["output_double_prime"].mean().reset_index()

# --- Compute bin widths and deltas ---
days_grouped["Œî Days"] = days_grouped["days_bin"].apply(lambda x: x.right - x.left)
days_grouped["Mean output_double_prime"] = days_grouped["output_double_prime"]
days_grouped["Œî Output"] = days_grouped["Mean output_double_prime"].diff()
days_grouped["$/day"] = (days_grouped["Œî Output"] / days_grouped["Œî Days"]).round(4)

# --- Output day rate table ---
days_df = days_grouped[["days_bin", "Mean output_double_prime", "Œî Output", "Œî Days", "$/day"]]
days_df.columns = ["Bucket", "Mean output_double_prime", "Œî Output", "Œî Days", "$/day"]

print("\nüìä Per-day reimbursement estimate per bin:")
print(days_df.round(4))

# --- Plot ---
days_grouped.set_index("days_bin")["Mean output_double_prime"].plot(
    kind="bar", figsize=(10, 4), title="output_double_prime by trip_duration_days buckets"
)
plt.ylabel("Mean output_double_prime")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# --- Final check: Reconstructed output vs actual ---
df["reconstructed"] = df["f_r"] + df["f_m"] + df["output_double_prime"]

sns.scatterplot(x=df["reconstructed"], y=df["expected_output"])
plt.title("Reconstructed vs Actual")
plt.xlabel("Reconstructed Output")
plt.ylabel("Expected Output")
plt.tight_layout()
plt.show()
```

```python
#### this script as a test harness to validate our reverse-engineered reimbursement policy: 
#### it applies the candidate f_final(days, miles, receipts)‚Äîwhich encodes our hypothesized caps, tiers, and rates‚Äîto each of the 
#### 1,000 public cases and compares the output to the known ‚Äúanswer key.‚Äù By counting mismatches and examining any discrepancies,
#### we can confirm whether our model perfectly replicates the legacy business logic or identify where our parameters or structure 
#### still need adjustment.
#### Just a test run to understand the problem before moving to hour 3.

import json
import pandas as pd

# Load data
with open("public_cases.json") as f:
    raw = json.load(f)

# Extract input and output
inputs = pd.json_normalize([item["input"] for item in raw])
expected = pd.Series([item["expected_output"] for item in raw])

# --- Final canonical function ---
def f_r(receipts):
    return min(receipts, 1800)

def f_m(miles):
    if miles <= 500:
        return 0
    elif miles <= 1000:
        return (miles - 500) * 0.41
    else:
        return 500 * 0.41 + (miles - 1000) * 0.20

def f_d(days):
    if days <= 7:
        return days * 75
    elif days <= 10:
        return 7 * 75
    else:
        return 7 * 75 + (days - 10) * 21

def f_final(days, miles, receipts):
    return round(f_r(receipts) + f_m(miles) + f_d(days), 2)

# --- Apply and compare ---
predicted = inputs.apply(
    lambda row: f_final(row["trip_duration_days"], row["miles_traveled"], row["total_receipts_amount"]),
    axis=1
)

# Validation
mismatches = predicted != expected
if mismatches.sum() == 0:
    print("‚úÖ PASS: f_final matches all public cases exactly.")
else:
    print(f"‚ùå FAIL: {mismatches.sum()} mismatches found.")
    print(inputs[mismatches])
    print("Details:")
    print(pd.DataFrame({
        "days": inputs["trip_duration_days"][mismatches],
        "miles": inputs["miles_traveled"][mismatches],
        "receipts": inputs["total_receipts_amount"][mismatches],
        "expected": expected[mismatches],
        "predicted": predicted[mismatches]
    }))
```

```python
# Our first model's failure proved that simply fitting data is the wrong approach. 
# We are now modeling the original business policy‚Äîwith its caps, tiers, and logic‚Äînot just statistical noise.
#
# Next iteration:
# This code implemented a second-generation model that abandoned the failed additive formula
# (f = receipts + mileage + per_diem) and instead tested a policy-based reimbursement structure. Specifically,
# it hypothesized that:
# Mileage is reimbursed via a tiered rate (with 3 bands: 0‚Äì500, 500‚Äì1000, 1000+).
# Receipts are reimbursed only up to a capped daily allowance, based on trip duration.

# The total reimbursement is therefore:
# f_policy_model = f_mileage(miles) + min(receipts, f_allowance(days)).

# The goal was to test whether this more realistic business logic could better match the
# 1,000 known examples‚Äîand it represented a fundamental shift from "fitting patterns" to reconstructing policy.

import json
import pandas as pd

# Load data
with open("public_cases.json") as f:
    raw = json.load(f)

# Extract input and output
inputs = pd.json_normalize([item["input"] for item in raw])
expected = pd.Series([item["expected_output"] for item in raw])

# --- New Component Hypotheses ---
# Note: These parameters are first-pass estimates and need tuning.

def f_mileage_new(miles):
    """
    Hypothesizes a tiered mileage reimbursement.
    The <500 mile tier is a placeholder guess to be refined.
    """
    if miles <= 500:
        # A simple low rate instead of zero
        rate1 = 0.10
        return miles * rate1
    elif miles <= 1000:
        # Rates derived from our granular `$/mile` analysis
        rate1 = 0.10
        rate2 = 0.41
        return (500 * rate1) + (miles - 500) * rate2
    else:
        rate1 = 0.10
        rate2 = 0.41
        rate3 = 0.20
        return (500 * rate1) + (500 * rate2) + (miles - 1000) * rate3

def f_allowance_new(days):
    """
    Hypothesizes a tiered daily allowance that acts as a CAP on receipts.
    This is the new interpretation of the per-diem component.
    """
    if days <= 7:
        # Rate derived from `$/day` analysis
        return days * 85
    elif days <= 10:
        # Models the 'plateau' identified in the analysis
        return 7 * 85
    else:
        # Models the 'bonus' for long trips
        return (7 * 85) + (days - 10) * 21

# --- Final Policy-Based Model ---
def f_policy_model(days, miles, receipts):
    """
    Implements the new thesis: Reimbursement = Mileage + min(Receipts, Allowance)
    This structure is fundamentally different from the previous additive model.
    """
    mileage_reimbursement = f_mileage_new(miles)
    daily_expense_reimbursement = min(receipts, f_allowance_new(days))
    
    return round(mileage_reimbursement + daily_expense_reimbursement, 2)

# --- Apply and compare ---
predicted = inputs.apply(
    lambda row: f_policy_model(row["trip_duration_days"], row["miles_traveled"], row["total_receipts_amount"]),
    axis=1
)

# Validation
mismatches = predicted != expected
if mismatches.sum() == 0:
    print("‚úÖ PASS: f_policy_model matches all public cases exactly.")
else:
    print(f"‚ùå FAIL: {mismatches.sum()} mismatches found.")
    print("Details of first 5 mismatches:")
    print(pd.DataFrame({
        "days": inputs["trip_duration_days"][mismatches].head(5),
        "miles": inputs["miles_traveled"][mismatches].head(5),
        "receipts": inputs["total_receipts_amount"][mismatches].head(5),
        "expected": expected[mismatches].head(5),
        "predicted": predicted[mismatches].head(5)
    }))
```
